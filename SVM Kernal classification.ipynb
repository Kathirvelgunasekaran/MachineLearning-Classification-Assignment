```python
#importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
#Always try to check and count no of rows and columns for this dataset.
```

```python
#Now we are going to convert categorial data column (Gender) as Numerical numbers using One Hot Encoding 
dataset=pd.get_dummies(dataset,drop_first=True) # Nominal data to adrenal data get, while giving drop_first=True function it'll drop extra column.
dataset
```

```python
dataset.columns #It'll show all dataset columns. we can use it for further process related.
```

```python
dataset=dataset.drop("User ID", axis=1)
```

```python
dataset["Purchased"].value_counts() #value_counts() means it'll show all unique entries of the dataset, and all unique entries based repeated counts like(0-257 and 1-143).
#Here we are finding how many of them have 0 and 1 entry,
#Classifictaion experts always first ask, 1.in our target varaiable"Purchased" how many class we are going to classify. 2. Data is balance or inbalanced.
#Balanced means data will be without partiality both will be like equal number(0 -250 and 1-250). balanced dataset means without patiality it'll learn(Ex: student 1 given 1/2 time to learn with 1 material, student 2 learn with 5 material and  day time provided, both of then not in balanced state.)
#For below statement(0 learn -257 entries and 1 learn -143 entries) 0 and 1 value have difference so it's called inbalance data. Each company have inbalanced data handle related have some parametersor percentage. if it's inbalanced within 5% means they said ok, if inbalanced data less than 5% means they will move to some othe stm, purely it's based on problem stmt and requirement.
```

```python
#Input and output split
#Data was already preprocessed. Now we are going to perform Input and output variables split
indep=dataset[['Age', 'EstimatedSalary', 'Gender_Male']] 
indep #Now 'Age', 'EstimatedSalary', 'Gender_Male' column values seperately assigned into independent varaiable, Independent means input, Dependent means output.
#Now 'charges' column value seperately assigned into dependent varaiable, dependent means output.
dep=dataset[["Purchased"]] #Now charges value seperately assigned into dependent varaiable,
dep
#Now we got seperatley independent and dependent varaiables. now sucessfully splited input and output
```

```python
indep.shape
```

```python
dep
```

```python
#Split Trainset and Testset. 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep, dep, test_size=1/3, random_state=0) #No varaiable assigned here.here parallely split input and output based on 70:30 ratio.
#we are using _(underscore) instead of space into vraiables(X_train,X_test,y_train,y_test)
#X means input, y means output, X input based on trainset and testset(X_train,X_test) and also y was output based trainset and testset(y_train,y_test) 
#if you want to use sklearn library based on trainset and testset you have to split it like this four variables(X_train,X_test,y_train,y_test).
#As of now compltetd Trainset and Testset split.
```

```python
#Model Creation
#ML algorithm all are provided by sklearn, we are creting this Model using trainset and tesset varaiable. 
from sklearn.svm import SVC # While calling, this line loaded into memory. here SVR may be function or class. In this line SVR considered as class
classifier=SVC(kernel='rbf', random_state=0) #Non parameterized function we are assigning into regressor varaiable.it consider as class, all the function under this class assign to regressor.
classifier=classifier.fit(X_train,y_train) #fit function we are calling it using access operator, now passing splited trainset input and output.
```

```python
#Evaluation metrix(R2 using)
y_pred=classifier.predict(X_test) #y-pred is varaiable, all are saved in regressor(formula,weight,bais), predict is application function.
#Evaluation matrix should happened on Testdataset .Anyone of the Testset values not included into Trandataset values. so that we are using it for crosss predition
#y-test is Actual data(y_test), X_test is brain predicted outptut.so that we are using it for crosss prediction
```

```python
#Classification means we have to use confusion_matrix
from sklearn.metrics import confusion_matrix #using confusion_matrix instead of r2_score function.
cm=confusion_matrix(y_test,y_pred) #confusion_matrix assigned to cm variable, now passing Actual output(y_test) and predicted output(y_pred)
```

```python
print(cm) #how to undertsand this going to next coding
```

```python
from sklearn.metrics import classification_report #using this confusion_matrix function going to find classification_report value.
clf_report=classification_report(y_test,y_pred) #under classification_report war against( Actual data(y_test) and Predicted data(y_pred))
#1.What is overall performance of DecisionTree- 0.68(68%)
#2.What kind of problem statement- Classification problem statement
#3.What is overall performance of your Model--Accuracy is overall percentage 0.68(68%)
#4.What is the Recall of 0(0 means 'not purchased' and 1 means 'purhased')-- 0.99 (99%)
#5.What is the correct performance of purchased? --0.14(14%)
#6.Purchased-Correctly classified as well as wrongly classified-->(Model telling about whole value including wrongly classified)--0.88(88%)
#7.What is the f1-score of 0- 0.80(80%)
#8.What is the f1-score of 1- 0.25(25%)
#9.MacroAvg value for Precision--0.77(77%)
#10.MacroAvg value for Recall--0.57(57%)
#11.MacroAvg value for F1-Score--0.52(52%)
#12.Weighted average value for Precision--0.74(74%)
#13.Weighted average value for Recall--0.68(68%)
#14.Weighted average value for F1-Score--0.59(59%)
#15.Support value(Testdata count) -134
#16.It's Balanced dataset/Imbalanced dataset--Imbalanced dataset
```

```python
print(clf_report) #with help of confusion_matrix you can answer nearly 7 questions(precision,recall,f-score,support)
#using confusion_matrix you can easily solve deep learning concepts. DL maximum classification stmt, where classifiction matrix is ther confusion_matrix will follow
```
